````python
# π› οΈ κ°„λ‹¨ν• RSA κµ¬ν„! (κµμ΅μ©, μ‹¤μ λ΅λ” λΌμ΄λΈλ¬λ¦¬ μ‚¬μ©)

import random
import math

print("π› οΈ κ°„λ‹¨ν• RSA κµ¬ν„!")
print("="*50)

# π” μ†μ νλ³„ λ°©λ²•λ“¤ λΉ„κµ!

print("π§® μ†μ νλ³„ κ³Όμ • μ‹¤μ λ΅ λ³΄κΈ°!")
print("="*50)

# 1. κΈ°λ³Έ μ†μ νλ³„ (λλ¦° λ°©λ²•)
def is_prime_basic_demo(n):
    """κΈ°λ³Έ λ°©λ²•: λ¨λ“  μλ΅ λ‚λ„μ–΄λ³΄κΈ° (κ³Όμ • ν‘μ‹)"""
    if n < 2:
        return False
    
    print(f"   π” {n}μ΄ μ†μμΈμ§€ ν™•μΈ μ¤‘...")
    checked = 0
    for i in range(2, int(math.sqrt(n)) + 1):
        checked += 1
        if n % i == 0:
            print(f"   β {i}λ΅ λ‚λ„μ–΄λ–¨μ–΄μ§ β†’ μ†μ μ•„λ‹ ({checked}λ² ν™•μΈ)")
            return False
    
    print(f"   β… λ¨λ“  μλ΅ μ• λ‚λ„μ–΄μ§ β†’ μ†μμ„! ({checked}λ² ν™•μΈ)")
    return True

# μ†μ νλ³„ ν…μ¤νΈ
test_numbers = [97, 100, 101]
for num in test_numbers:
    print(f"\nπ”Ά {num} ν…μ¤νΈ:")
    result = is_prime_basic_demo(num)

print(f"\nπ’΅ ν° μμ—μ„λ”?")
print(f"   617μλ¦¬ μ νλ³„: β(617μλ¦¬) β‰ 308μλ¦¬ μκΉμ§€ ν™•μΈ")
print(f"   ν™•μΈν•΄μ•Ό ν•  μμ κ°μ: μ—„μ²­λ‚κ² λ§μ!")
print(f"   β†’ κΈ°λ³Έ λ°©λ²•μΌλ΅λ” λ¶κ°€λ¥!")

print(f"\nβ΅ μ‹¤μ λ΅λ” 'λ°€λ¬-λΌλΉ ν…μ¤νΈ' μ‚¬μ©:")
print(f"   π² ν™•λ¥ μ  λ°©λ²• (99.999% μ •ν™•)")
print(f"   β° λ§¤μ° λΉ λ¦„ (λ‡ μ΄)")
print(f"   π”„ μ—¬λ¬ λ² λ°λ³µν•΄μ„ ν™•μ‹¤μ„± λ†’μ„")

# 1. κ°„λ‹¨ν• μ†μ μ°ΎκΈ° ν•¨μ
def is_prime(n):
    """κ°„λ‹¨ν• μ†μ νλ³„ (μ‘μ€ μμ©)"""
    if n < 2:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True

def find_prime(start, end):
    """λ²”μ„μ—μ„ μ†μ μ°ΎκΈ°"""
    for n in range(start, end):
        if is_prime(n):
            return n
    return None

# 2. ν™•μ¥ μ ν΄λ¦¬λ“ νΈμ λ²• (λ¨λ“λ¬ μ—­μ› μ°ΎκΈ°)
def extended_gcd(a, b):
    if a == 0:
        return b, 0, 1
    gcd, x1, y1 = extended_gcd(b % a, a)
    x = y1 - (b // a) * x1
    y = x1
    return gcd, x, y

def mod_inverse(e, phi):
    """eμ Ο†(n)μ— λ€ν• λ¨λ“λ¬ μ—­μ› μ°ΎκΈ°"""
    gcd, x, _ = extended_gcd(e, phi)
    if gcd != 1:
        raise Exception("λ¨λ“λ¬ μ—­μ›μ΄ μ΅΄μ¬ν•μ§€ μ•μ")
    return (x % phi + phi) % phi

# 3. RSA ν‚¤ μƒμ„± (μ‘μ€ μλ΅ λ°λ¨)
print("π”‘ RSA ν‚¤ μƒμ„± κ³Όμ •:")

# μ‹¤μ  RSA ν‚¤ μƒμ„± κ³Όμ • μ‹λ®¬λ μ΄μ…
print("π² μ»΄ν“¨ν„°κ°€ λλ¤μΌλ΅ μ†μ μƒμ„±ν•λ” κ³Όμ •:")
print("-" * 50)

def generate_random_prime(min_val, max_val):
    """λλ¤ν•κ² μ†μλ¥Ό μ°Ύλ” κ³Όμ • μ‹λ®¬λ μ΄μ…"""
    import random
    attempts = 0
    while True:
        attempts += 1
        candidate = random.randint(min_val, max_val)
        if is_prime(candidate):
            print(f"   μ‹λ„ {attempts}λ²μ§Έ: {candidate} β†’ μ†μ λ°κ²¬! β…")
            return candidate
        else:
            print(f"   μ‹λ„ {attempts}λ²μ§Έ: {candidate} β†’ μ†μ μ•„λ‹ β")
        
        if attempts >= 5:  # λ°λ¨μ©μΌλ΅ 5λ²λ§ λ³΄μ—¬μ£ΌκΈ°
            print(f"   ... (μ‹¤μ λ΅λ” κ³„μ† λ°λ³µ)")
            break
    
    # λ°λ¨μ©μΌλ΅ λ―Έλ¦¬ μ•κ³  μλ” μ†μ λ°ν™
    return find_prime(min_val, max_val)

print("π” μ²« λ²μ§Έ μ†μ p μ°Ύλ” μ¤‘...")
p = generate_random_prime(50, 100)

print("\nπ” λ‘ λ²μ§Έ μ†μ q μ°Ύλ” μ¤‘...")  
q = generate_random_prime(100, 150)

print(f"\nβ… μƒμ„± μ™„λ£!")
print(f"   μ†μ p = {p}")
print(f"   μ†μ q = {q}")
print(f"   β†’ μ‚¬μ©μλ” μ΄ κ³Όμ •μ„ μ „ν€ λ¨λ¦„!")
print(f"   β†’ μ»΄ν“¨ν„°κ°€ μλ™μΌλ΅ λλ¤ μƒμ„±!")

print(f"   1. μ†μ p = {p}")
print(f"   2. μ†μ q = {q}")

# nκ³Ό Ο†(n) κ³„μ‚°
n = p * q
phi_n = (p - 1) * (q - 1)

print(f"   3. n = p Γ— q = {n}")
print(f"   4. Ο†(n) = (p-1) Γ— (q-1) = {phi_n}")

# e μ„ νƒ (65537μ΄ ν‘μ¤€μ΄μ§€λ§ μ‘μ€ μμ μ©μΌλ΅ 3 μ‚¬μ©)
e = 3
while math.gcd(e, phi_n) != 1:
    e += 2  # ν™€μλ΅ μ¦κ°€

# d κ³„μ‚° (eμ λ¨λ“λ¬ μ—­μ›)
d = mod_inverse(e, phi_n)

print(f"   5. e = {e}")
print(f"   6. d = {d}")

print(f"\nπ”‘ μƒμ„±λ ν‚¤:")
print(f"   π“Ά κ³µκ°ν‚¤: (n={n}, e={e})")
print(f"   π” κ°μΈν‚¤: (n={n}, d={d})")

# 4. RSA μ•”νΈν™”/λ³µνΈν™” ν•¨μ
def rsa_encrypt(message, n, e):
    """RSA μ•”νΈν™”"""
    return pow(message, e, n)

def rsa_decrypt(ciphertext, n, d):
    """RSA λ³µνΈν™”"""
    return pow(ciphertext, d, n)

# 5. ν…μ¤νΈ!
message = 42  # μ›λ³Έ λ©”μ‹μ§€ (μ«μ)
print(f"\nπ§ RSA μ•”νΈν™”/λ³µνΈν™” ν…μ¤νΈ:")
print(f"   π“ μ›λ³Έ λ©”μ‹μ§€: {message}")

# μ•”νΈν™” (κ³µκ°ν‚¤ μ‚¬μ©)
encrypted = rsa_encrypt(message, n, e)
print(f"   π” μ•”νΈν™”: {message}^{e} mod {n} = {encrypted}")

# λ³µνΈν™” (κ°μΈν‚¤ μ‚¬μ©)
decrypted = rsa_decrypt(encrypted, n, d)
print(f"   π”“ λ³µνΈν™”: {encrypted}^{d} mod {n} = {decrypted}")

print(f"\nβ… κ²°κ³Ό: {message} β†’ {encrypted} β†’ {decrypted}")
print(f"μ„±κ³µ! μ›λ³Έκ³Ό λ³µνΈν™” κ²°κ³Όκ°€ κ°™μ: {message == decrypted}")

print(f"\nπ’΅ ν•µμ‹¬ ν¬μΈνΈ:")
print(f"   π”‘ κ³µκ°ν‚¤λ΅ μ•”νΈν™”, κ°μΈν‚¤λ΅ λ³µνΈν™”")
print(f"   π§® κ±°λ“­μ κ³± μ—°μ‚°μ λ§λ²•!")
print(f"   π”’ nμ„ μ†μΈμλ¶„ν•΄ν•  μ μ—†μ–΄μ•Ό μ•μ „!")
````
