## π― λ©ν‘: "λ€μΉ­μ•”νΈκ°€ λ­”μ§€λ§ μ•κ³  λ„μ–΄κ°€μ!"

**ZKP κ°€λ” κΈΈμ— μ κΉ λ“¤λ¬κ°€λ” μ •λ„!** πƒβ€β™‚οΈ

### π“ ν•µμ‹¬λ§ λΉ λ¥΄κ²:
- **λ€μΉ­μ•”νΈλ€?** μ•”νΈν™”/λ³µνΈν™”μ— κ°™μ€ ν‚¤ μ‚¬μ©
- **AESκ°€ λ­”κ°€?** ν„μ¬ ν‘μ¤€ λ€μΉ­μ•”νΈ
- **ECB vs CBC** λ¨λ“ μ°¨μ΄μ λ§ λ€μ¶©
- **μ–Έμ  μ“°λ‚?** λΉ λ¥Έ λ°μ΄ν„° μ•”νΈν™”

## π¤” λ€μΉ­μ•”νΈ = κ°™μ€ μ—΄μ‡ 

**μ‰½κ² λ§ν•΄μ„**: μ κ·Έλ” μ—΄μ‡  = μ—¬λ” μ—΄μ‡  π”‘

```
ν‰λ¬Έ + λΉ„λ°€ν‚¤ β†’ [AES μ•”νΈν™”] β†’ μ•”νΈλ¬Έ
μ•”νΈλ¬Έ + λΉ„λ°€ν‚¤ β†’ [AES λ³µνΈν™”] β†’ ν‰λ¬Έ
```

**μ¥μ **: λΉ λ¦„ β΅  
**λ‹¨μ **: ν‚¤ κ³µμ  λ¬Έμ  π° (μ–΄λ–»κ² μ•μ „ν•κ² ν‚¤λ¥Ό μ „λ‹¬?)

````python
# π”’ λ€μΉ­μ•”νΈ κ°λ…λ§ λΉ λ¥΄κ²! (λΌμ΄λΈλ¬λ¦¬ μ—†μ΄ μ‹λ®¬λ μ΄μ…)
import random

print("π”’ λ€μΉ­μ•”νΈ λΉ λ¥Έ μ΄ν•΄!")
print("="*40)

# 1. λ€μΉ­μ•”νΈ κ°λ… μ‹λ®¬λ μ΄μ… (μ‹¤μ  AESλ” μ•„λ‹!)
def simple_encrypt(message, key):
    """κ°„λ‹¨ν• XOR μ•”νΈν™” (κ°λ… μ΄ν•΄μ©)"""
    encrypted = []
    for i, char in enumerate(message):
        encrypted.append(ord(char) ^ key)
    return encrypted

def simple_decrypt(encrypted, key):
    """κ°„λ‹¨ν• XOR λ³µνΈν™” (κ°λ… μ΄ν•΄μ©)"""
    decrypted = ""
    for num in encrypted:
        decrypted += chr(num ^ key)
    return decrypted

# 2. ν‚¤ μƒμ„± (λλ¤)
secret_key = random.randint(1, 255)
print(f"π”‘ μƒμ„±λ λΉ„λ°€ν‚¤: {secret_key}")

# 3. μ•”νΈν™”
message = "λΉ„λ°€λ©”μ‹μ§€!"
encrypted = simple_encrypt(message, secret_key)

print(f"π“ μ›λ³Έ λ©”μ‹μ§€: {message}")
print(f"π” μ•”νΈν™”λ λ©”μ‹μ§€: {encrypted[:5]}... (μ«μλ“¤)")

# 4. λ³µνΈν™” (κ°™μ€ ν‚¤ μ‚¬μ©!)
decrypted = simple_decrypt(encrypted, secret_key)
print(f"π”“ λ³µνΈν™”λ λ©”μ‹μ§€: {decrypted}")

# 5. ν•µμ‹¬ ν¬μΈνΈ
print(f"\nπ’΅ ν•µμ‹¬ ν¬μΈνΈ:")
print(f"   β… κ°™μ€ ν‚¤λ΅ μ•”νΈν™”/λ³µνΈν™” (ν‚¤: {secret_key})")
print(f"   β΅ μ‹¤μ  AESλ” ν›¨μ”¬ λ³µμ΅ν•κ³  μ•μ „ν•¨")
print(f"   π° λ¬Έμ : ν‚¤λ¥Ό μ–΄λ–»κ² μ•μ „ν•κ² κ³µμ ?")

# 6. ν‚¤κ°€ λ‹¤λ¥΄λ©΄?
print(f"\nπ¨ ν‚¤κ°€ λ‹¤λ¥΄λ©΄ μ–΄λ–»κ² λ κΉ?")
wrong_key = random.randint(1, 255)
wrong_decrypted = simple_decrypt(encrypted, wrong_key)

print(f"   β λ‹¤λ¥Έ ν‚¤({wrong_key})λ΅ λ³µνΈν™”: {wrong_decrypted} (κΉ¨μ§!)")
    
print(f"\nπ― λ€μΉ­μ•”νΈ μ”μ•½: κ°™μ€ ν‚¤, λΉ λ¥Έ μ†λ„, ν‚¤ κ³µμ  λ¬Έμ !")
print(f"π’΅ μ‹¤μ λ΅λ” AES-256 κ°™μ€ μ•μ „ν• μ•κ³ λ¦¬μ¦ μ‚¬μ©!")
````

````python
# π― λ€μΉ­μ•”νΈ vs κ³µκ°ν‚¤ μ•”νΈ λΉ„κµ!

print("π― λ€μΉ­μ•”νΈ vs κ³µκ°ν‚¤ μ•”νΈ!")
print("="*50)

# μ‹λ®¬λ μ΄μ… (μ‹¤μ λ΅λ” κµ¬ν„ μ• ν•¨)
print("π”’ λ€μΉ­μ•”νΈ (AES):")
print("   π”‘ ν‚¤: 1κ° (λΉ„λ°€ν‚¤)")  
print("   β΅ μ†λ„: λ§¤μ° λΉ λ¦„")
print("   π“¦ μ©λ„: λ€μ©λ‰ λ°μ΄ν„°")
print("   π° λ¬Έμ : ν‚¤ κ³µμ κ°€ μ–΄λ ¤μ›€")
print("   π’΅ μμ‹: νμΌ μ•”νΈν™”, VPN")

print("\nπ”“ κ³µκ°ν‚¤ μ•”νΈ (RSA):")
print("   π”‘ ν‚¤: 2κ° (κ³µκ°ν‚¤ + κ°μΈν‚¤)")
print("   π μ†λ„: λλ¦Ό")  
print("   π“¦ μ©λ„: μ‘μ€ λ°μ΄ν„°, ν‚¤ κµν™")
print("   β… μ¥μ : ν‚¤ κ³µμ  λ¬Έμ  ν•΄κ²°!")
print("   π’΅ μμ‹: HTTPS, λ””μ§€ν„Έ μ„λ…")

print("\nπ¤ μ‹¤μ λ΅λ” ν•¨κ» μ‚¬μ©:")
print("   1. RSAλ΅ AES ν‚¤λ¥Ό μ•μ „ν•κ² μ „λ‹¬")
print("   2. AESλ΅ μ‹¤μ  λ°μ΄ν„°λ¥Ό λΉ λ¥΄κ² μ•”νΈν™”")
print("   β†’ λ‘μ μ¥μ μ„ λ¨λ‘ ν™μ©! π‰")

print("\nπ€ ZKPλ΅ κ°€κΈ° μ„ν•΄ μ•μ•„μ•Ό ν•  κ²ƒ:")
print("   β… λ€μΉ­μ•”νΈ: κ°™μ€ ν‚¤, λΉ λ¦„, ν‚¤ κ³µμ  λ¬Έμ ")
print("   π”„ λ‹¤μ: κ³µκ°ν‚¤ μ•”νΈ (RSA κΈ°μ΄)")
print("   π― μµμΆ…: ZKP (Zero-Knowledge Proof)")

print("\nπ“ λ€μΉ­μ•”νΈ μ™„λ£! λ‹¤μ λ‹¨κ³„λ΅ γ„±γ„±!")
````
