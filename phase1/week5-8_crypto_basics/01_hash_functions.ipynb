## âš”ï¸ SHA-256 vs MD5 ëŒ€ê²°!

### ğŸ¥Š MD5 (êµ¬ì‹ ì±”í”¼ì–¸)
```
- ì¶œë ¥: 128ë¹„íŠ¸ (32ì)
```

### ğŸ† SHA-256 (í˜„ì¬ ì±”í”¼ì–¸)
```
- ì¶œë ¥: 256ë¹„íŠ¸ (64ì)  
- ì†ë„: ë³´í†µ ğŸš€
- ë³´ì•ˆ: ê°•ë ¥í•¨ âœ… (í˜„ì¬ê¹Œì§€ ì•ˆì „)
- ìš©ë„: ë¸”ë¡ì²´ì¸, íŒ¨ìŠ¤ì›Œë“œ, ë””ì§€í„¸ ì„œëª…
```

**ğŸ¯ ê²°ë¡ : SHA-256 ì“°ì„¸ìš”!** MD5ëŠ” ì´ì œ ì€í‡´...


````python
# âš”ï¸ SHA-256 vs MD5 ì‹¤ì œ ëŒ€ê²°!
import time

print("âš”ï¸ SHA-256 vs MD5 ëŒ€ê²°!")
print("="*50)

test_message = "ì´ê²ƒì€ í•´ì‹œ í•¨ìˆ˜ í…ŒìŠ¤íŠ¸ ë©”ì‹œì§€ì…ë‹ˆë‹¤!" * 1000  # ê¸´ ë©”ì‹œì§€

print(f"ğŸ“ í…ŒìŠ¤íŠ¸ ë©”ì‹œì§€ ê¸¸ì´: {len(test_message):,} ë¬¸ì")
print()

# MD5 í…ŒìŠ¤íŠ¸
print("ğŸ¥Š MD5 (êµ¬ì‹ ì±”í”¼ì–¸):")
start_time = time.time()
md5_hash = hashlib.md5(test_message.encode('utf-8')).hexdigest()
md5_time = time.time() - start_time

print(f"   ê²°ê³¼: {md5_hash}")
print(f"   ê¸¸ì´: {len(md5_hash)} ë¬¸ì (128ë¹„íŠ¸)")
print(f"   ì‹œê°„: {md5_time:.6f}ì´ˆ")
print()

# SHA-256 í…ŒìŠ¤íŠ¸  
print("ğŸ† SHA-256 (í˜„ì¬ ì±”í”¼ì–¸):")
start_time = time.time()
sha256_hash = hashlib.sha256(test_message.encode('utf-8')).hexdigest()
sha256_time = time.time() - start_time

print(f"   ê²°ê³¼: {sha256_hash}")
print(f"   ê¸¸ì´: {len(sha256_hash)} ë¬¸ì (256ë¹„íŠ¸)")
print(f"   ì‹œê°„: {sha256_time:.6f}ì´ˆ")
print()

# ì†ë„ ë¹„êµ
speed_ratio = sha256_time / md5_time if md5_time > 0 else 1
print(f"âš¡ ì†ë„ ë¹„êµ:")
print(f"   MD5ê°€ SHA-256ë³´ë‹¤ {speed_ratio:.1f}ë°° ë¹ ë¦„")
print(f"   í•˜ì§€ë§Œ ë³´ì•ˆì€ SHA-256ì´ í›¨ì”¬ ì•ˆì „! ğŸ›¡ï¸")

print(f"\nğŸ¯ ê²°ë¡ :")
print(f"   ğŸ“ˆ ì†ë„: MD5 ìŠ¹ë¦¬")
print(f"   ğŸ›¡ï¸ ë³´ì•ˆ: SHA-256 ì••ìŠ¹")
print(f"   ğŸ† ì‹¤ì „: SHA-256 ì‚¬ìš© ì¶”ì²œ!")
````

## ğŸ¯ ì‹¤ì „ í™œìš©: ê°„ë‹¨í•œ ë¸”ë¡ì²´ì¸ ë§Œë“¤ê¸°!

**ë¸”ë¡ì²´ì¸ì˜ í•µì‹¬ = í•´ì‹œ ì²´ì¸!** 

ê° ë¸”ë¡ì´ ì´ì „ ë¸”ë¡ì˜ í•´ì‹œë¥¼ í¬í•¨í•´ì„œ ì²´ì¸ì²˜ëŸ¼ ì—°ê²°ë©ë‹ˆë‹¤.

```
Block 1: [Data] â†’ Hash1
Block 2: [Data + Hash1] â†’ Hash2  
Block 3: [Data + Hash2] â†’ Hash3
```

**ë§Œì•½ Block 1 ë°ì´í„°ê°€ ë°”ë€Œë©´?**
â†’ Hash1 ë°”ë€œ â†’ Hash2 ë°”ë€œ â†’ Hash3 ë°”ë€œ â†’ **ì „ë¶€ ë‹¤ ë°”ë€œ!**

**â†’ ì´ë˜ì„œ ë¸”ë¡ì²´ì¸ì´ ìœ„ì¡° ë¶ˆê°€ëŠ¥í•œê±°ì˜ˆìš”!** ğŸ”’


````python
# ğŸ”— ê°„ë‹¨í•œ ë¸”ë¡ì²´ì¸ êµ¬í˜„!
import json
from datetime import datetime

class SimpleBlock:
    def __init__(self, data, previous_hash="0"):
        self.timestamp = datetime.now().isoformat()
        self.data = data
        self.previous_hash = previous_hash
        self.hash = self.calculate_hash()
    
    def calculate_hash(self):
        # ë¸”ë¡ì˜ ëª¨ë“  ì •ë³´ë¥¼ í•©ì³ì„œ í•´ì‹œ ê³„ì‚°
        block_string = f"{self.timestamp}{self.data}{self.previous_hash}"
        return hashlib.sha256(block_string.encode()).hexdigest()
    
    def __str__(self):
        return f"Block(data='{self.data}', hash='{self.hash[:16]}...')"

class SimpleBlockchain:
    def __init__(self):
        self.chain = [self.create_genesis_block()]
    
    def create_genesis_block(self):
        return SimpleBlock("Genesis Block", "0")
    
    def get_latest_block(self):
        return self.chain[-1]
    
    def add_block(self, data):
        previous_block = self.get_latest_block()
        new_block = SimpleBlock(data, previous_block.hash)
        self.chain.append(new_block)
    
    def is_chain_valid(self):
        for i in range(1, len(self.chain)):
            current_block = self.chain[i]
            previous_block = self.chain[i-1]
            
            # í˜„ì¬ ë¸”ë¡ì˜ í•´ì‹œê°€ ì˜¬ë°”ë¥¸ì§€ í™•ì¸
            if current_block.hash != current_block.calculate_hash():
                return False
            
            # ì´ì „ ë¸”ë¡ê³¼ì˜ ì—°ê²°ì´ ì˜¬ë°”ë¥¸ì§€ í™•ì¸
            if current_block.previous_hash != previous_block.hash:
                return False
        
        return True

print("ğŸ”— ê°„ë‹¨í•œ ë¸”ë¡ì²´ì¸ ë§Œë“¤ê¸°!")
print("="*50)

# ë¸”ë¡ì²´ì¸ ìƒì„±
blockchain = SimpleBlockchain()

# ë¸”ë¡ ì¶”ê°€
print("ğŸ“¦ ë¸”ë¡ë“¤ ì¶”ê°€ ì¤‘...")
blockchain.add_block("ì†¡ê¸ˆ: ì² ìˆ˜ â†’ ì˜í¬ 100ì›")
blockchain.add_block("ì†¡ê¸ˆ: ì˜í¬ â†’ ë¯¼ìˆ˜ 50ì›")
blockchain.add_block("ì†¡ê¸ˆ: ë¯¼ìˆ˜ â†’ ì² ìˆ˜ 25ì›")

# ë¸”ë¡ì²´ì¸ ìƒíƒœ ì¶œë ¥
print(f"\nğŸ” ë¸”ë¡ì²´ì¸ ìƒíƒœ:")
for i, block in enumerate(blockchain.chain):
    print(f"Block {i}: {block}")
    print(f"   ë°ì´í„°: {block.data}")
    print(f"   ì´ì „ í•´ì‹œ: {block.previous_hash[:16] if block.previous_hash != '0' else '0'}...")
    print(f"   í˜„ì¬ í•´ì‹œ: {block.hash[:16]}...")
    print()

# ë¸”ë¡ì²´ì¸ ìœ íš¨ì„± ê²€ì¦
is_valid = blockchain.is_chain_valid()
print(f"âœ… ë¸”ë¡ì²´ì¸ ìœ íš¨ì„±: {'ìœ íš¨í•¨' if is_valid else 'ë¬´íš¨í•¨'}")

# í•´í‚¹ ì‹œë®¬ë ˆì´ì…˜!
print(f"\nğŸ”´ í•´í‚¹ ì‹œë®¬ë ˆì´ì…˜: Block 1 ë°ì´í„° ë³€ì¡°!")
blockchain.chain[1].data = "ì†¡ê¸ˆ: ì² ìˆ˜ â†’ ì˜í¬ 1000ì›"  # 100ì› â†’ 1000ì›ìœ¼ë¡œ ë³€ì¡°!

is_valid_after_hack = blockchain.is_chain_valid()
print(f"ğŸš¨ ë³€ì¡° í›„ ë¸”ë¡ì²´ì¸ ìœ íš¨ì„±: {'ìœ íš¨í•¨' if is_valid_after_hack else 'ë¬´íš¨í•¨ (í•´í‚¹ ê°ì§€!)'}")

print(f"\nğŸ’¡ í•´ì‹œ ì²´ì¸ì˜ í˜:")
print(f"   - í•˜ë‚˜ì˜ ë¸”ë¡ë§Œ ë°”ë€Œì–´ë„ ì¦‰ì‹œ ê°ì§€ ê°€ëŠ¥!")
print(f"   - ì´ê²ƒì´ ë¸”ë¡ì²´ì¸ì´ ìœ„ì¡° ë¶ˆê°€ëŠ¥í•œ ì´ìœ ! ğŸ”’")
````

````python
# ğŸ¯ ZKPë¥¼ ìœ„í•œ í•µì‹¬ ê°œë…ë“¤ ì‹¤ìŠµ!

print("ğŸ¯ ZKPë¥¼ ìœ„í•œ í•´ì‹œ í•µì‹¬ ê°œë…!")
print("="*50)

# 1. Commitment Scheme (ì•½ì† ë°©ì‹)
print("ğŸ’ 1. Commitment Scheme ì‹¤ìŠµ:")
import secrets

secret_value = "1234"  # ë‚´ê°€ ì•Œê³  ìˆëŠ” ë¹„ë°€
nonce = secrets.token_hex(16)  # ëœë¤ ê°’

# ì•½ì† ë‹¨ê³„: í•´ì‹œë¡œ "ë´‰ì¸"
commitment = hashlib.sha256((secret_value + nonce).encode()).hexdigest()
print(f"   ë¹„ë°€ê°’: {secret_value}")
print(f"   ë…¼ìŠ¤: {nonce}")
print(f"   ì•½ì†(í•´ì‹œ): {commitment[:32]}...")

# ê³µê°œ ë‹¨ê³„: ë¹„ë°€ê³¼ ë…¼ìŠ¤ë¥¼ ê³µê°œí•´ì„œ ê²€ì¦
verification = hashlib.sha256((secret_value + nonce).encode()).hexdigest()
is_valid = commitment == verification
print(f"   ê²€ì¦ ê²°ê³¼: {'ì„±ê³µ! ì•½ì† ì§€ì¼œì§' if is_valid else 'ì‹¤íŒ¨! ì•½ì† ìœ„ë°˜'} âœ…")

print()

# 2. Saltì˜ ì¤‘ìš”ì„±
print("ğŸ§‚ 2. Saltì˜ ì¤‘ìš”ì„±:")
password = "password123"

# Salt ì—†ì´
no_salt_hash1 = hashlib.sha256(password.encode()).hexdigest()
no_salt_hash2 = hashlib.sha256(password.encode()).hexdigest()

# Salt ìˆì´
salt1 = secrets.token_hex(8)
salt2 = secrets.token_hex(8)
with_salt_hash1 = hashlib.sha256((password + salt1).encode()).hexdigest()
with_salt_hash2 = hashlib.sha256((password + salt2).encode()).hexdigest()

print(f"   ê°™ì€ ë¹„ë°€ë²ˆí˜¸ '{password}':")
print(f"   Salt ì—†ì´: {no_salt_hash1[:32]}... (í•­ìƒ ë™ì¼)")
print(f"   Salt ì—†ì´: {no_salt_hash2[:32]}... (í•­ìƒ ë™ì¼)")
print(f"   Salt ìˆì´: {with_salt_hash1[:32]}... (ë§¤ë²ˆ ë‹¤ë¦„)")
print(f"   Salt ìˆì´: {with_salt_hash2[:32]}... (ë§¤ë²ˆ ë‹¤ë¦„)")

print()

# 3. í•´ì‹œ ì²´ì¸ (ZKP ì¦ëª…ì— í™œìš©)
print("ğŸ”— 3. í•´ì‹œ ì²´ì¸ ì¦ëª…:")
original_secret = "my_secret"

# í•´ì‹œ ì²´ì¸ ìƒì„± (10ë‹¨ê³„)
chain = [original_secret]
for i in range(10):
    last_hash = hashlib.sha256(chain[-1].encode()).hexdigest()
    chain.append(last_hash)

print(f"   ì›ë³¸ ë¹„ë°€: {original_secret}")
print(f"   1ë‹¨ê³„: {chain[1][:32]}...")
print(f"   5ë‹¨ê³„: {chain[5][:32]}...")
print(f"   10ë‹¨ê³„: {chain[10][:32]}...")

# ì²´ì¸ ê²€ì¦ (ê±°ê¾¸ë¡œ í™•ì¸)
print(f"   â†’ 10ë‹¨ê³„ë¶€í„° ê±°ê¾¸ë¡œ ê²€ì¦í•˜ë©´ ì›ë³¸ê¹Œì§€ ì¦ëª… ê°€ëŠ¥!")

print()
print("ğŸ“ í•µì‹¬ í¬ì¸íŠ¸:")
print("   1. Commitment: ë¹„ë°€ì„ ë¨¼ì € ì•½ì†í•˜ê³  ë‚˜ì¤‘ì— ê³µê°œ")
print("   2. Salt/Nonce: ê°™ì€ ê°’ë„ ë§¤ë²ˆ ë‹¤ë¥´ê²Œ ë³´ì´ë„ë¡")
print("   3. í•´ì‹œ ì²´ì¸: ë‹¨ê³„ë³„ ì¦ëª…ì— í™œìš©")
print("   4. ì´ ëª¨ë“  ê²Œ ZKPì˜ ê¸°ì´ˆê°€ ë©ë‹ˆë‹¤! ğŸš€")
````
