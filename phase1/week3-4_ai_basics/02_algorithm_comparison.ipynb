# ğŸ¥Š AI ì•Œê³ ë¦¬ì¦˜ ëŒ€ê²°! Decision Tree vs Random Forest vs SVM

## ğŸ¤– **ì¶œì „ ì„ ìˆ˜ë“¤**

### 1ï¸âƒ£ **Decision Tree** (ì˜ì‚¬ê²°ì •ë‚˜ë¬´) ğŸŒ³
- **íŠ¹ì§•**: ë‚˜ë¬´ì²˜ëŸ¼ ì§ˆë¬¸í•˜ë©´ì„œ ë‹µ ì°¾ê¸°
- **ì¥ì **: ì´í•´í•˜ê¸° ì‰¬ì›€
- **ë‹¨ì **: ê³¼ì í•© ìœ„í—˜

### 2ï¸âƒ£ **Random Forest** (ëœë¤ ìˆ²) ğŸŒ³ğŸŒ³ğŸŒ³  
- **íŠ¹ì§•**: ë‚˜ë¬´ ì—¬ëŸ¬ ê°œë¥¼ í•©ì³ì„œ íˆ¬í‘œ
- **ì¥ì **: ë” ì •í™•í•¨
- **ë‹¨ì **: ì´í•´í•˜ê¸° ì–´ë ¤ì›€

### 3ï¸âƒ£ **SVM** (Support Vector Machine) âš”ï¸
- **íŠ¹ì§•**: ê²½ê³„ì„ ì„ ê·¸ì–´ì„œ êµ¬ë¶„
- **ì¥ì **: ë³µì¡í•œ íŒ¨í„´ë„ ì°¾ìŒ

## ğŸ¯ **ëŒ€ê²° ë°©ì‹**
**ê°™ì€ ì•„ì´ë¦¬ìŠ¤ ê½ƒ ë°ì´í„°**ë¡œ **ì •í™•ë„ ë¹„êµ**!


```python
# ğŸ§° ëŒ€ê²°ì— í•„ìš”í•œ ë„êµ¬ë“¤ ì¤€ë¹„
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report

# 3ëª…ì˜ ì¶œì „ ì„ ìˆ˜ë“¤!
from sklearn.tree import DecisionTreeClassifier           # ğŸŒ³ ë‚˜ë¬´
from sklearn.ensemble import RandomForestClassifier      # ğŸŒ³ğŸŒ³ğŸŒ³ ìˆ²  
from sklearn.svm import SVC                              # âš”ï¸ SVM

print("ğŸ¥Š AI ëŒ€ê²° ì¤€ë¹„ ì™„ë£Œ!")
print("âœ… Decision Tree - ë‚˜ë¬´ ì „ì‚¬")
print("âœ… Random Forest - ìˆ² êµ°ë‹¨") 
print("âœ… SVM - ê²½ê³„ì„  ë§ˆìŠ¤í„°")
print("\nâš”ï¸ ëŒ€ê²° ì‹œì‘í•©ë‹ˆë‹¤!")
```

```python
# ğŸ² ëŒ€ê²° ë°ì´í„° ì¤€ë¹„ (ì•„ì´ë¦¬ìŠ¤ ê½ƒ)
print("ğŸŒ¸ ëŒ€ê²°ì¥ ì¤€ë¹„: ì•„ì´ë¦¬ìŠ¤ ê½ƒ ë°ì´í„°!")
print("="*50)

# ë°ì´í„° ë¡œë“œ
iris = datasets.load_iris()
X = iris.data   # ê½ƒ í¬ê¸° (4ê°œ íŠ¹ì„±)
y = iris.target # ê½ƒ ì¢…ë¥˜ (3ê°œ í´ë˜ìŠ¤)

print(f"ğŸ“Š ëŒ€ê²° ë°ì´í„°:")
print(f"   ê½ƒ ê°œìˆ˜: {len(X)}ê°œ")
print(f"   ì¸¡ì • íŠ¹ì„±: {len(iris.feature_names)}ê°œ")
print(f"   ê½ƒ ì¢…ë¥˜: {len(iris.target_names)}ê°œ")

# ğŸ”¥ ëŒ€ê²°ì„ ì–´ë µê²Œ! í›ˆë ¨ ë°ì´í„° ëŒ€í­ ì¤„ì´ê¸°!
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.7, random_state=42  # 30%ë§Œ í•™ìŠµ, 70% í…ŒìŠ¤íŠ¸!
)

print(f"\nâš–ï¸ ğŸ”¥ í•˜ë“œì½”ì–´ ëŒ€ê²° í™˜ê²½:")
print(f"   í›ˆë ¨ìš©: {len(X_train)}ê°œ (ì ì€ í•™ìŠµ ë°ì´í„°!)")
print(f"   í…ŒìŠ¤íŠ¸ìš©: {len(X_test)}ê°œ (ë§ì€ í‰ê°€ ë°ì´í„°!)")
print(f"\nğŸ’€ ë‚œì´ë„: HARD")
print("   - ì ì€ ë°ì´í„°ë¡œ í•™ìŠµí•´ì•¼ í•¨")
print("   - ê³¼ì í•© ìœ„í—˜ ì¦ê°€") 
print("   - ì•Œê³ ë¦¬ì¦˜ ì§„ì§œ ì‹¤ë ¥ í…ŒìŠ¤íŠ¸!")
print("\nğŸŸï¸ í•˜ë“œì½”ì–´ ëŒ€ê²°ì¥ ì¤€ë¹„ ì™„ë£Œ!")
```

```python
# ğŸ¥Š 1ë¼ìš´ë“œ: 3ëª…ì˜ AI ì„ ìˆ˜ í›ˆë ¨ì‹œí‚¤ê¸°!
print("ğŸ‹ï¸ 1ë¼ìš´ë“œ: í›ˆë ¨ ì‹œê°„!")
print("="*50)

# ì„ ìˆ˜ 1: Decision Tree ğŸŒ³
print("ğŸŒ³ Decision Tree í›ˆë ¨ ì¤‘...")
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)
print("   âœ… ë‚˜ë¬´ ì „ì‚¬ í›ˆë ¨ ì™„ë£Œ!")

# ì„ ìˆ˜ 2: Random Forest ğŸŒ³ğŸŒ³ğŸŒ³
print("\nğŸŒ³ğŸŒ³ğŸŒ³ Random Forest í›ˆë ¨ ì¤‘...")
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)  # ë‚˜ë¬´ 100ê°œ!
rf_model.fit(X_train, y_train)
print("   âœ… ìˆ² êµ°ë‹¨ í›ˆë ¨ ì™„ë£Œ! (ë‚˜ë¬´ 100ê°œ ë™ì›)")

# ì„ ìˆ˜ 3: SVM âš”ï¸
print("\nâš”ï¸ SVM í›ˆë ¨ ì¤‘...")
svm_model = SVC(random_state=42)
svm_model.fit(X_train, y_train)
print("   âœ… ê²½ê³„ì„  ë§ˆìŠ¤í„° í›ˆë ¨ ì™„ë£Œ!")

print("\nğŸ¯ ëª¨ë“  ì„ ìˆ˜ í›ˆë ¨ ì™„ë£Œ! ì´ì œ ì‹¤ë ¥ í…ŒìŠ¤íŠ¸!")
```

```python
# ğŸ† 2ë¼ìš´ë“œ: ìµœì¢… ëŒ€ê²°! ì •í™•ë„ ë¹„êµ
print("ğŸ† 2ë¼ìš´ë“œ: ìµœì¢… ëŒ€ê²°!")
print("="*50)

# ê° ì„ ìˆ˜ì˜ ì˜ˆì¸¡ ê²°ê³¼
dt_pred = dt_model.predict(X_test)
rf_pred = rf_model.predict(X_test)
svm_pred = svm_model.predict(X_test)

# ì •í™•ë„ ê³„ì‚°
dt_accuracy = accuracy_score(y_test, dt_pred)
rf_accuracy = accuracy_score(y_test, rf_pred)
svm_accuracy = accuracy_score(y_test, svm_pred)

print("ğŸ“Š ëŒ€ê²° ê²°ê³¼:")
print(f"ğŸŒ³ Decision Tree:    {dt_accuracy:.1%}")
print(f"ğŸŒ³ğŸŒ³ğŸŒ³ Random Forest:   {rf_accuracy:.1%}")
print(f"âš”ï¸ SVM:              {svm_accuracy:.1%}")

# ìš°ìŠ¹ì ë°œí‘œ!
accuracies = [dt_accuracy, rf_accuracy, svm_accuracy]
names = ["Decision Tree ğŸŒ³", "Random Forest ğŸŒ³ğŸŒ³ğŸŒ³", "SVM âš”ï¸"]
winner_idx = accuracies.index(max(accuracies))

print(f"\nğŸ¥‡ ìš°ìŠ¹ì: {names[winner_idx]}")
print(f"ğŸ‰ ìµœê³  ì •í™•ë„: {max(accuracies):.1%}")

# ìì„¸í•œ ë¶„ì„ (ìš°ìŠ¹ìë§Œ)
print(f"\nğŸ“ˆ ìš°ìŠ¹ì ìƒì„¸ ë¶„ì„:")
if winner_idx == 0:
    winner_pred = dt_pred
elif winner_idx == 1:
    winner_pred = rf_pred
else:
    winner_pred = svm_pred

print(classification_report(y_test, winner_pred, target_names=iris.target_names))
```

```python
# ğŸ§ª ë³´ë„ˆìŠ¤: ìƒˆë¡œìš´ ê½ƒìœ¼ë¡œ 3ëª… ëª¨ë‘ í…ŒìŠ¤íŠ¸!
print("ğŸ§ª ë³´ë„ˆìŠ¤ ë¼ìš´ë“œ: ìƒˆë¡œìš´ ê½ƒ ì˜ˆì¸¡ ëŒ€ê²°!")
print("="*50)

# ìƒˆë¡œìš´ ê°€ìƒì˜ ê½ƒ 3ê°œ
test_flowers = [
    [5.1, 3.5, 1.4, 0.2],  # ì‘ì€ ê½ƒ (setosa ê°™ìŒ)
    [6.0, 2.9, 4.5, 1.5],  # ì¤‘ê°„ ê½ƒ (versicolor ê°™ìŒ)  
    [7.2, 3.6, 6.1, 2.5]   # í° ê½ƒ (virginica ê°™ìŒ)
]

flower_descriptions = ["ì‘ì€ ê½ƒ", "ì¤‘ê°„ ê½ƒ", "í° ê½ƒ"]

for i, (flower, desc) in enumerate(zip(test_flowers, flower_descriptions)):
    print(f"\nğŸŒº {i+1}ë²ˆì§¸ í…ŒìŠ¤íŠ¸: {desc}")
    print(f"   í¬ê¸°: {flower}")
    
    # 3ëª…ì˜ ì˜ˆì¸¡
    dt_pred = dt_model.predict([flower])[0]
    rf_pred = rf_model.predict([flower])[0] 
    svm_pred = svm_model.predict([flower])[0]
    
    print(f"   ğŸŒ³ Decision Tree:  {iris.target_names[dt_pred]}")
    print(f"   ğŸŒ³ğŸŒ³ğŸŒ³ Random Forest: {iris.target_names[rf_pred]}")
    print(f"   âš”ï¸ SVM:            {iris.target_names[svm_pred]}")
    
    # ì˜ê²¬ ì¼ì¹˜ë„ í™•ì¸
    predictions = [dt_pred, rf_pred, svm_pred]
    if len(set(predictions)) == 1:
        print("   âœ… ëª¨ë“  AIê°€ ë™ì˜!")
    else:
        print("   ğŸ¤” AIë“¤ ì˜ê²¬ì´ ë‹¤ë¦„...")

print(f"\nğŸŠ 3ê°œ ì•Œê³ ë¦¬ì¦˜ ë¹„êµ ì™„ë£Œ!")
print("ğŸ“š ë°°ìš´ ê²ƒ:")
print("- Decision Tree: ì´í•´í•˜ê¸° ì‰½ì§€ë§Œ ê³¼ì í•© ìœ„í—˜")
print("- Random Forest: ì—¬ëŸ¬ ë‚˜ë¬´ë¡œ ë” ì•ˆì •ì ")  
print("- SVM: ë³µì¡í•œ ê²½ê³„ì„ ìœ¼ë¡œ ì •êµí•œ ë¶„ë¥˜")
print("\nğŸš€ ë‹¤ìŒ: íšŒê·€ ì•Œê³ ë¦¬ì¦˜ìœ¼ë¡œ ìˆ«ì ì˜ˆì¸¡í•´ë³´ê¸°!")
```

## ğŸ§ª íŠ¹ë³„ ì‹¤í—˜: ë°ì´í„° ë¶€ì¡± ì‹œë®¬ë ˆì´ì…˜!

**ì§ˆë¬¸**: í›ˆë ¨ ë°ì´í„°ê°€ ì ì  ì¤„ì–´ë“¤ë©´ ì–´ë–»ê²Œ ë ê¹Œ? ğŸ¤”

**ê°€ì„¤**: ë°ì´í„°ê°€ ì ì„ìˆ˜ë¡ ì„±ëŠ¥ì´ ë–¨ì–´ì§€ê³  ì•Œê³ ë¦¬ì¦˜ ì°¨ì´ê°€ ê·¹ëª…í•´ì§ˆ ê²ƒ!

**ì‹¤í—˜**: test_sizeë¥¼ 0.3 â†’ 0.7 â†’ 0.8 â†’ 0.9ë¡œ ì ì§„ì ìœ¼ë¡œ ëŠ˜ë ¤ë³´ê¸°


```python
# ğŸ”¬ ë°ì´í„° ë¶€ì¡± ì‹¤í—˜: ê·¹í•œê¹Œì§€ ë°€ì–´ë¶™ì´ê¸°!
print("ğŸ”¬ ë°ì´í„° ë¶€ì¡± ì‹¤í—˜ ì‹œì‘!")
print("="*60)

# ë‹¤ì–‘í•œ test_sizeë¡œ ì‹¤í—˜ (í›ˆë ¨ ë°ì´í„°ê°€ ì ì  ì¤„ì–´ë“¦)
test_sizes = [0.3, 0.7, 0.8, 0.9]
experiment_names = [
    "0.3 (105ê°œ í•™ìŠµ, 45ê°œ í…ŒìŠ¤íŠ¸) - ì¼ë°˜ì ",
    "0.7 (45ê°œ í•™ìŠµ, 105ê°œ í…ŒìŠ¤íŠ¸) - ì–´ë ¤ì›€", 
    "0.8 (30ê°œ í•™ìŠµ, 120ê°œ í…ŒìŠ¤íŠ¸) - ë§¤ìš° ì–´ë ¤ì›€",
    "0.9 (15ê°œ í•™ìŠµ, 135ê°œ í…ŒìŠ¤íŠ¸) - ê·¹í•œ!"
]

results = []

for test_size, exp_name in zip(test_sizes, experiment_names):
    print(f"\nğŸ§ª ì‹¤í—˜: {exp_name}")
    print("-" * 50)
    
    # ë°ì´í„° ë¶„í• 
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, random_state=42
    )
    
    print(f"ğŸ“Š ì‹¤ì œ ë°ì´í„°: í›ˆë ¨ {len(X_train)}ê°œ, í…ŒìŠ¤íŠ¸ {len(X_test)}ê°œ")
    
    # 3ê°œ ëª¨ë¸ í›ˆë ¨
    dt_model = DecisionTreeClassifier(random_state=42).fit(X_train, y_train)
    rf_model = RandomForestClassifier(random_state=42, n_estimators=100).fit(X_train, y_train)
    svm_model = SVC(random_state=42).fit(X_train, y_train)
    
    # ì„±ëŠ¥ ì¸¡ì •
    dt_acc = accuracy_score(y_test, dt_model.predict(X_test))
    rf_acc = accuracy_score(y_test, rf_model.predict(X_test))
    svm_acc = accuracy_score(y_test, svm_model.predict(X_test))
    
    print(f"ğŸŒ³ Decision Tree:  {dt_acc:.1%}")
    print(f"ğŸŒ³ğŸŒ³ğŸŒ³ Random Forest: {rf_acc:.1%}")
    print(f"âš”ï¸ SVM:            {svm_acc:.1%}")
    
    # ìš°ìŠ¹ì ì°¾ê¸° (ë™ì  ì²˜ë¦¬ í¬í•¨)
    accuracies = [dt_acc, rf_acc, svm_acc]
    algorithm_names = ["Decision Tree ğŸŒ³", "Random Forest ğŸŒ³ğŸŒ³ğŸŒ³", "SVM âš”ï¸"]
    max_accuracy = max(accuracies)
    
    # ìµœê³  ì ìˆ˜ë¥¼ ë°›ì€ ëª¨ë“  ì•Œê³ ë¦¬ì¦˜ ì°¾ê¸°
    winners = [algorithm_names[i] for i, acc in enumerate(accuracies) if acc == max_accuracy]
    
    if len(winners) == 1:
        winner_text = winners[0]
        print(f"ğŸ¥‡ ì´ ë¼ìš´ë“œ ìš°ìŠ¹ì: {winner_text} ({max_accuracy:.1%})")
    else:
        winner_text = " & ".join(winners)
        print(f"ğŸ¥‡ ì´ ë¼ìš´ë“œ ìš°ìŠ¹ì: {winner_text} (ê³µë™ìš°ìŠ¹ {max_accuracy:.1%})")
    
    # ê²°ê³¼ ì €ì¥
    results.append({
        'test_size': test_size,
        'train_count': len(X_train),
        'dt_acc': dt_acc,
        'rf_acc': rf_acc, 
        'svm_acc': svm_acc,
        'winner': winner_text
    })

print(f"\nğŸŠ ëª¨ë“  ì‹¤í—˜ ì™„ë£Œ!")
```

```python
# ğŸ“Š ì‹¤í—˜ ê²°ê³¼ ë¶„ì„ ë° ë†€ë¼ìš´ ë°œê²¬!
print("ğŸ“Š ì‹¤í—˜ ê²°ê³¼ ì¢…í•© ë¶„ì„")
print("="*60)

# ê²°ê³¼ í‘œë¡œ ì •ë¦¬
print("\nğŸ“‹ ì „ì²´ ê²°ê³¼ ìš”ì•½:")
print(f"{'í›ˆë ¨ ë°ì´í„°':<8} {'Decision Tree':<15} {'Random Forest':<15} {'SVM':<15} {'ìš°ìŠ¹ì':<20}")
print("-" * 75)

for result in results:
    print(f"{result['train_count']:>6}ê°œ   "
          f"{result['dt_acc']:>12.1%}   "
          f"{result['rf_acc']:>13.1%}   "
          f"{result['svm_acc']:>11.1%}   "
          f"{result['winner']:<20}")

print(f"\nğŸ¤¯ ë†€ë¼ìš´ ë°œê²¬ë“¤:")
print("="*40)

print("1ï¸âƒ£ **ì˜ˆìƒê³¼ ì™„ì „ ë°˜ëŒ€!**")
print("   - ê°€ì„¤: 'ë°ì´í„° ì ì„ìˆ˜ë¡ ì„±ëŠ¥ ë–¨ì–´ì§ˆ ê²ƒ'")
print("   - ì‹¤ì œ: 'ë°ì´í„° ì ì–´ë„ ì„±ëŠ¥ ìœ ì§€/í–¥ìƒ!'")

print("\n2ï¸âƒ£ **ì•Œê³ ë¦¬ì¦˜ ìˆœìœ„ ë³€í™”!**")
print("   - 105ê°œ í•™ìŠµ: ëª¨ë‘ 100% (êµ¬ë¶„ ë¶ˆê°€)")  
print("   - 45ê°œ í•™ìŠµ: SVM ìš°ìŠ¹ (96.2%)")
print("   - 30ê°œ í•™ìŠµ: ëª¨ë‘ ë™ì  (93.3%)")
print("   - 15ê°œ í•™ìŠµ: Decision Tree ì—­ì „ìŠ¹! (95.6%)")

print("\n3ï¸âƒ£ **ì•„ì´ë¦¬ìŠ¤ ë°ì´í„°ì˜ íŠ¹ì„±**")
print("   - ê½ƒ ì¢…ë¥˜ê°€ ëª…í™•íˆ êµ¬ë¶„ë¨")
print("   - ì ì€ ë°ì´í„°ë¡œë„ ì¶©ë¶„íˆ í•™ìŠµ ê°€ëŠ¥")
print("   - 'ê°„ë‹¨í•œ ë¬¸ì œ'ì˜ ëŒ€í‘œ ì‚¬ë¡€")

print("\nğŸ¯ **êµí›ˆ**:")
print("âœ… ì•Œê³ ë¦¬ì¦˜ ì„±ëŠ¥ì€ ë°ì´í„° ì–‘ì— ë”°ë¼ ë‹¬ë¼ì§")
print("âœ… ë³µì¡í•œ ì•Œê³ ë¦¬ì¦˜ì´ í•­ìƒ ì¢‹ì€ ê±´ ì•„ë‹˜")  
print("âœ… ë¬¸ì œê°€ ë‹¨ìˆœí•˜ë©´ ë‹¨ìˆœí•œ ë°©ë²•ì´ ë” íš¨ê³¼ì ")
print("âœ… ì‹¤í—˜ì„ í†µí•´ì„œë§Œ ì•Œ ìˆ˜ ìˆëŠ” ê²ƒë“¤ì´ ìˆìŒ!")

print(f"\nğŸ”¬ **ê³¼í•™ì  ì ‘ê·¼ì˜ ì¤‘ìš”ì„±**:")
print("ê°€ì„¤ â†’ ì‹¤í—˜ â†’ ê´€ì°° â†’ ë¶„ì„ â†’ ìƒˆë¡œìš´ ë°œê²¬! ğŸ§ªâœ¨")
```
